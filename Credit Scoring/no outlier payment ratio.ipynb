```python
import matplotlib.pyplot as plt
def nooutlier(data,name):
    data1 = data[data.default_flag == 1][[name,'default_flag']]
    data0 = data[data.default_flag == 0][[name,'default_flag']]
    Q1def = data1[name].quantile(0.25)
    Q3def = data1[name].quantile(0.75)
    IQRdef = Q3def - Q1def
    data1noout = data1[(data1[name] > (Q1def - 1.5 * IQRdef)) & (data1[name] < (Q3def + 1.5 * IQRdef))]
    
    Q1nodef = data0[name].quantile(0.25)
    Q3nodef = data0[name].quantile(0.75)
    IQRnodef = Q3nodef - Q1nodef
    data0noout = data0[(data0[name] > (Q1nodef - 1.5 * IQRnodef)) & (data0[name] < (Q3nodef + 1.5 * IQRnodef))]
    datanoout = pd.concat([data1noout,data0noout],axis = 0)
    return sns.boxplot(y = name,x = 'default_flag',data=datanoout)
```

```python
nooutlier(data,'payment_ratio_3month')
plt.title('Payment Ratio (3 months) from majority of clients\ndefault vs not default')
plt.xlabel('Default Category')
plt.ylabel('Payment Ratio (3 months)')
```
