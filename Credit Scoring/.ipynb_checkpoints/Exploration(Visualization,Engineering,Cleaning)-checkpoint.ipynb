```python
data = pd.read_csv('credit_ds_v3.csv')
data.drop('X',axis = 1,inplace=True)
data.info()
data.head()
```

```python
#check for weird data
print(data[data.outstanding > data.credit_limit][['outstanding','remaining_bill','credit_limit','overlimit_percentage']])

print(data[data.overlimit_percentage > 0][['outstanding','remaining_bill','credit_limit','overlimit_percentage']])

print(data.overlimit_percentage.describe())

print(data[data.overlimit_percentage  == 190][['outstanding','remaining_bill','credit_limit','overlimit_percentage']])

print(data[data.overlimit_percentage != data.remaining_bill_per_limit][['remaining_bill_per_limit','remaining_bill','credit_limit','overlimit_percentage']])
#also check for total_usage_per_limit, just in case
print(data[['total_usage','credit_limit','total_usage_per_limit']].head(10))
```

26,45% (4138/15645) of our data is a person who used their credit card way past their limits. It is quite unclear on how to get the value of overlimit percentage. It should either be by (outstanding-limit/limit) or (remaining_bill-limit/limit). remaining_bill_per_limit is already available as a variable.
We will redefine "overlimit_percentage" as (outstanding-limit/limit). By definition, outstanding is the amount of loan that hasn't been paid, so it will somehow describe the amount of funds over the credit limit.
(https://www.lawinsider.com/dictionary/outstanding-loan)

```python
data.overlimit_percentage = ((data.outstanding - data.credit_limit)/data.credit_limit)*100
data.overlimit_percentage = np.where(data.overlimit_percentage > 0,data.overlimit_percentage, 0)
print(data.overlimit_percentage.describe())
data[data.overlimit_percentage >599][['outstanding','credit_limit','overlimit_percentage','remaining_bill_per_limit','default_flag']]
```

This is weird. Outstanding is way above credit_limit yet the default_flag is 0

```python
data.branch_code.value_counts()
toreduce = data.drop(['branch_code','default_flag'],axis = 1,inplace=False)
```

```python
#let's proceed with data reduction due to the high number of variables that correlates one another
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
pcascale = StandardScaler()
forpca = pd.DataFrame(pcascale.fit_transform(toreduce),columns=toreduce.columns)
forpca.dropna(inplace=True)
pca = PCA()
reduced = pca.fit_transform(forpca)
reduced = pd.DataFrame(reduced)
reduced.head()
```

```python
def display_scree_plot(pca):
    scree = pca.explained_variance_ratio_*100
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="red",marker='o')
    plt.xlabel("Number of Principal Components")
    plt.ylabel("Percentage Explained Variance")
    plt.title("Scree Plot")
    plt.show(block=False)
```

```python
import numpy as np
expvar = np.round(pca.explained_variance_ratio_,5) 
expvar
print(pca.explained_variance_ratio_.cumsum())
display_scree_plot(pca) 
#only need 12 variables to have 90.57% of the variance explained
```

```python
loadings = pd.DataFrame(pca.components_.T, columns=['PC1', 'PC2','PC3','PC4',
                                                    'PC5','PC6','PC7','PC8',
                                                    'PC9','PC10','PC11','PC12',
                                                    'PC13','PC14','PC15','PC16',
                                                    'PC17','PC18','PC19','PC20',
                                                    'PC21'], index=forpca.columns).iloc[:,:12]
loadings[(loadings>0.3)|(loadings<-0.3)]

#we can compute the 'loadings matrix' to directly understand correlation between
#variables and principal component but 'loadings' can actually describe the correlation 
#between variables and principal components indirectly.
#we can choose variables with the near highest absolute loading to represent a component

#we choose:
#outstanding, utilization_3month, total_usage_per_limit,
#payment_ratio_3month,number_of_cards,total_3mo_usage_per_limit,
#total_cash_usage,delinquency_score,payment_ratio_6month,
#years_since_card_issuing,overlimit_percentage,payment_ratio
```

What if we use Information Value to select which feature is useful to the classification?

```python
x = data.drop('default_flag',axis = 1,inplace = False)
y = data['default_flag']
```

```python
#using package from https://pypi.org/project/xverse/
from xverse.transformer import WOE

FS = WOE()
FS.fit(x, y)
IV = FS.iv_df
IV[IV.Information_Value>0.1]
```

I refer to this link to understand select which feature is important:
https://towardsdatascience.com/intro-to-credit-scorecard-9afeaaa3725f

it has the same result with PCA, more or less. We'll continue with calculating the credit score of each clients using what variable is available. We'll use the variables within Information Value since it shows us how relevant a variable to the default_flag variable, unlike PCA.

```python
datatouse = data[IV[IV.Information_Value > 0.1].Variable_Name.tolist()]
datatouse['default_flag'] = data.default_flag
print(datatouse.corr()[(datatouse.corr()>0.3)|(datatouse.corr()<-0.3)])
#let's drop remaining bill per limit, total_retail usage, and utilization_6month
datatouse.drop(['remaining_bill_per_limit', 'total_retail_usage', 'utilization_6month'],axis = 1,inplace=True)
datatouse.info()
```

```python
#to calculate credit score as additional variable
#I have no prior knowledge on calculating credit score so I'll try my best to present it well
#I refer to this link to calculate credit score and to choose which variables will be used
#https://www.wellsfargo.com/financial-education/credit-management/calculate-credit-score/

#remaining bill -> amount owed (0.3 multiplier) -> lower the better
#payment_ratio -> payment history (0.35 multiplier) ->  higher the better
#overlimit_percentage -> amount owed (0.3 multiplier) -> lower the better
#payment_ratio_6month -> payment history (0.35 multiplier) -> higher the better
#delinquency_score -> payment history (0.35 multiplier) -> lower the better
#total_usage_per_limit -> recent credit activity/new credit (0.1 multiplier) -> lower the better
#total_3mo_usage_per_limit -> recent credit activity/new credit (0.1 multiplier) -> lower the better

#should we bring all score to the same level? Is it logical to use minmaxscaler?
#we expect that minmaxscaler would convert all value to be between 0 and 1
from sklearn.preprocessing import MinMaxScaler
normalize = MinMaxScaler()
csvar = datatouse[['remaining_bill','payment_ratio',
                   'overlimit_percentage','payment_ratio_6month',
                  'delinquency_score','total_usage_per_limit',
                  'total_3mo_usage_per_limit']]
csvar.isnull().sum()
```

```python
csvar.fillna(0,inplace=True)#this means we see missing delinquency_score as not known to be doing anything delinquent
forcs = pd.DataFrame(normalize.fit_transform(csvar),columns = csvar.columns)
datatouse['credit_score'] = (forcs.remaining_bill*-0.3 + forcs.payment_ratio*0.35 + forcs.overlimit_percentage*-0.3 + forcs.payment_ratio_6month*0.35 + forcs.delinquency_score*-0.35 + forcs.total_usage_per_limit*-0.1 +forcs.total_3mo_usage_per_limit*-0.1)*1000

sns.histplot(datatouse.credit_score,kde=True)
plt.title('Credit Score Distribution')
plt.show()
print(datatouse.credit_score.describe())

plt.figure()
sns.boxplot(datatouse.default_flag,datatouse.credit_score)
plt.title('Default and No Default Credit Score Distribution')
plt.show()
```

```python
print(datatouse[datatouse.credit_score<0]['default_flag'].value_counts())
#we can see that the majority of people with credit score lower than 0 tends to default
#we can see consider the negative value of credit score as 0, meaning when a credit score is 0, a person would very 
#much likely to default
datatouse.credit_score = np.where(datatouse.credit_score<0,0,datatouse.credit_score)
print(datatouse.credit_score.describe())
```

We can see that credit score tends to be low when a person defaults but it isn't clear where the credit score cutoff between default and no default would be.
