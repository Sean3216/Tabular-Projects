We will be using K-Prototype to cluster the data since the data consists of categorical and numerical data. We reduce the variables first to minimize the number of variables we'll be interpreting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('readytoclust.csv')
```

```python
data.info()
```

```python
datanum = data.loc[:,data.dtypes != 'object']
datanum.shape
```

```python
#reduce numeric variables to several small variables
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
pcascale = StandardScaler()
datanum = data.loc[:,data.dtypes != 'object']
datacat = data.loc[:,data.dtypes == 'object']
forpca = pd.DataFrame(pcascale.fit_transform(datanum),columns=datanum.columns)
pca = PCA(svd_solver = 'full')
pca.fit(forpca)
```

```python
def display_scree_plot(pca):
    scree = pca.explained_variance_ratio_*100
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(),c="red",marker='o')
    plt.xlabel("Number of Principal Components")
    plt.ylabel("Percentage Explained Variance")
    plt.title("Scree Plot")
    plt.show(block=False)
```

```python
expvar = np.round(pca.explained_variance_ratio_,5) 
expvar
print(pca.explained_variance_ratio_.cumsum())
display_scree_plot(pca) 
#I want to retain at least 80% of  the original data variance
#we need 7 variables to have 80.19% of the variance explained
```

```python
pcanew = PCA(n_components = 7,svd_solver = 'full')
pcanew.fit(forpca)
loadings = pd.DataFrame(pcanew.components_.T, columns=['PC1', 'PC2','PC3','PC4',
                                                       'PC5','PC6','PC7'], index=forpca.columns)
loadings[(loadings>0.3)|(loadings<-0.3)]
```

```python
#just to help interpretation
#PC2
print(datanum[['Teenhome','NumDealsPurchases','NumWebPurchases']].corr())

#PC3
print(datanum[['days_since_join','Age']].corr())

#PC4
print(datanum[['days_since_join','NumDealsPurchases','Kidhome','Age']].corr())

#PC6
print(datanum[['Kidhome','MntMeatProducts','MntGoldProds','NumDealsPurchases',
               'NumCatalogPurchases','days_since_join']].corr())
```

```python
reduced = pd.DataFrame(pcanew.transform(forpca),
                       columns = ['wealthiness','usedeals','relative_age',
                                  'relative_timesincejoin','relative_mntgoods',
                                  'activitymeasurement','goldstosweets'])
forclust = pd.concat([reduced.reset_index(drop=True),datacat.reset_index(drop=True)],axis = 1)
```

```python
# Function for plotting elbow curve
from kmodes.kprototypes import KPrototypes
Poscatcol = [forclust.columns.get_loc(i) for i in forclust.loc[:, forclust.dtypes == 'object'].columns]
cost = []
for i in range(2,7):
    clust = KPrototypes(n_clusters = i,init='Huang',random_state=42)
    clust.fit_predict(forclust,categorical = Poscatcol)
    cost.append(clust.cost_)
    
sns.lineplot(x = range(2,7),y = cost,marker = "+",dashes = True)
plt.title('Elbow Curve')
plt.xlabel('N of Clusters')
plt.ylabel('Cost')
plt.plot
```

```python
#I decide to cluster the data into 3 cluster based on the Elbow Curve.
#more cluster might be better but sometimes we don't want to group our customers into many segments since it will also
#give ourselves a hard time.
finalclust = KPrototypes(n_clusters = 3, init='Huang',random_state=42)
finalclust.fit_predict(forclust,categorical = [7, 8, 9, 10, 11, 12, 13, 14, 15])
full = forclust.loc[:]
full['segments'] = finalclust.labels_
full.segments.value_counts()
```

```python
#let's plot the numerical variables first
plt.style.use('ggplot') 
fig, ax = plt.subplots(nrows = 3, ncols = 3,figsize = (10,10))
sns.boxplot(y='wealthiness', x= 'segments', data=full, ax=ax[0,0])
sns.boxplot(y='usedeals', x= 'segments', data=full, ax=ax[0,1])
sns.boxplot(y='relative_age', x= 'segments', data=full, ax=ax[0,2])
sns.boxplot(y='relative_timesincejoin', x= 'segments', data=full, ax=ax[1,0])
sns.boxplot(y='relative_mntgoods', x= 'segments', data=full, ax=ax[1,1])
sns.boxplot(y='activitymeasurement', x= 'segments', data=full, ax=ax[1,2])
sns.boxplot(y='goldstosweets', x= 'segments', data=full, ax=ax[2,0])
fig.delaxes(ax[2,1])
fig.delaxes(ax[2,2])
plt.tight_layout()
plt.show()
```

```python
#let's try interpreting the categorical variables
Eduplot = full.groupby('segments')['Education'].value_counts(normalize=True)
Eduplot = Eduplot.mul(100).rename('Percent').reset_index()

Maritalplot = full.groupby('segments')['Marital_Status'].value_counts(normalize=True)
Maritalplot = Maritalplot.mul(100).rename('Percent').reset_index()

Response = full.groupby('segments')['Response'].value_counts(normalize=True)
Response = Response.mul(100).rename('Percent').reset_index()

Complain = full.groupby('segments')['Complain'].value_counts(normalize=True)
Complain = Complain.mul(100).rename('Percent').reset_index()

Cmp1 = full.groupby('segments')['AcceptedCmp1'].value_counts(normalize=True)
Cmp1 = Cmp1.mul(100).rename('Percent').reset_index()

Cmp2 = full.groupby('segments')['AcceptedCmp2'].value_counts(normalize=True)
Cmp2 = Cmp2.mul(100).rename('Percent').reset_index()

Cmp3 = full.groupby('segments')['AcceptedCmp3'].value_counts(normalize=True)
Cmp3 = Cmp3.mul(100).rename('Percent').reset_index()

Cmp4 = full.groupby('segments')['AcceptedCmp4'].value_counts(normalize=True)
Cmp4 = Cmp4.mul(100).rename('Percent').reset_index()

Cmp5 = full.groupby('segments')['AcceptedCmp5'].value_counts(normalize=True)
Cmp5 = Cmp5.mul(100).rename('Percent').reset_index()

fig, ax = plt.subplots(nrows = 3, ncols = 4,figsize = (20,12))
sns.barplot(x='Education',y='Percent', hue= 'segments', data=Eduplot, ax=ax[0,0],dodge=True,order = ['Basic','2n Cycle','Graduation','Master','PhD'])
sns.barplot(x='Marital_Status',y='Percent', hue= 'segments', data=Maritalplot, ax=ax[0,1],dodge=True)
sns.barplot(x='Response',y='Percent', hue= 'segments', data=Response, ax=ax[0,2],dodge=True)
sns.barplot(x='Complain',y='Percent', hue= 'segments', data=Complain, ax=ax[0,3],dodge=True)
sns.barplot(x='AcceptedCmp1',y='Percent', hue= 'segments', data=Cmp1, ax=ax[1,0],dodge=True)
sns.barplot(x='AcceptedCmp2',y='Percent', hue= 'segments', data=Cmp2, ax=ax[1,1],dodge=True)
sns.barplot(x='AcceptedCmp3',y='Percent', hue= 'segments', data=Cmp3, ax=ax[1,2],dodge=True)
sns.barplot(x='AcceptedCmp4',y='Percent', hue= 'segments', data=Cmp4, ax=ax[1,3],dodge=True)
sns.barplot(x='AcceptedCmp5',y='Percent', hue= 'segments', data=Cmp5, ax=ax[2,0],dodge=True)
fig.delaxes(ax[2,1])
fig.delaxes(ax[2,2])
fig.delaxes(ax[2,3])
plt.tight_layout()
plt.show()
```
