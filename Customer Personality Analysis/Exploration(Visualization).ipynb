```python
#Correlation between amount spent on each items (Wines, Fruits, Meat, Fish, Sweets, and Gold products)
print('\033[1m'+'Correlation of Spendings'+'\033[0m')
sns.pairplot(data[['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']])
plt.show()
print('We can see that correlation between several variables is present but it is hard to interpret them.\n\n\n\n\n\n\n\n\n\n')


#Correlation between places of purchase (NumWebPurchases, NumCatalogPurchases, NumStorePurchases, NumWebVisitsMonth)
print('\033[1m'+'Check for correlation between web purchases and web visitation first'+'\033[1m')
sns.scatterplot(x='NumWebPurchases',y='NumWebVisitsMonth',data = data)
plt.show()
print(data[['NumWebPurchases','NumWebVisitsMonth']].corr())
print('We can see that the correlation is actually very weak\n\n')

print('\033[1m'+'Correlation of Places to Purchase'+'\033[0m')
sns.pairplot(data[['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']])
plt.show()
print('Correlation also exist within the place of purchase variables.')
print(data[['NumWebPurchases','NumCatalogPurchases']].corr())
print('We should also try clustering using PAM again to check for hidden groups of customer by their place of purchases. \n\n\n\n\n\n\n\n\n\n')


#Checking dates
print('\033[1m'+'Check for Dt_Customer'+'\033[0m')

print(data.Dt_Customer.describe())

```

```python
#Data Cleaning
#so there were a couple of things that we would want to clean first before clustering. We also wanted to create a
#new variable from the dt_customer

#Handling weird outliers
firstdrop = data[data.Income < 400000][:]
seconddrop = firstdrop[firstdrop.Year_Birth > 1935][:]
final = seconddrop[seconddrop.NumWebPurchases < 20][:]

#Handling date customer first
import datetime

final.Dt_Customer = pd.to_datetime(final.Dt_Customer)
final['days_since_join'] = (max(final.Dt_Customer)-final.Dt_Customer).dt.days
final['Age'] = (max(final.Dt_Customer.dt.year)-final.Year_Birth)


#Handling marital status
final['Marital_Status'] = final['Marital_Status'].replace({'Alone':'Single', 'Absurd':'Single','YOLO':'Single'})

#Check for missing values
print(final.isnull().sum())

#replace acceptedcampaign, complain and response
final[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Complain','Response']] = final[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Complain','Response']].replace(0,'No')
final[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Complain','Response']] = final[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Complain','Response']].replace(1,'Yes')
```
