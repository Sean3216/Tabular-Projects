We will be using K-Prototype to cluster the data since the data consists of categorical and numerical data. We reduce the variables first to minimize the number of variables we'll be interpreting.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('readytoclust.csv')
```

```python
data.info()
```

```python
datanum = data.loc[:,data.dtypes != 'object']
datanum.shape
```

```python
#reduce numeric variables to several small variables
from sklearn.preprocessing import StandardScaler
from factor_analyzer import FactorAnalyzer,calculate_bartlett_sphericity,calculate_kmo
pcascale = StandardScaler()
datanum = data.loc[:,data.dtypes != 'object']
datacat = data.loc[:,data.dtypes == 'object']
forpca = pd.DataFrame(pcascale.fit_transform(datanum),columns=datanum.columns)
forpca.dropna(inplace=True)
statbartlett,pvalbartlett = calculate_bartlett_sphericity(datanum)
kmo, kmoall = calculate_kmo(datanum)
print('Bartlett Sphericity Test:{0}\nKMO Test: {1}'.format(pvalbartlett,kmoall))
pca = FactorAnalyzer(n_factors = 16, rotation = None,method='principal')
pca.fit(forpca)
print('This would be the original Eigen Values:\n{0}'.format(pca.get_eigenvalues()[0]))
print("Based on Eigen Values, test suggests that 3 variables is enough.\nNow let's go check for the explained variances")
```

```python
def display_scree_plot(pca):
    scree = pca.get_factor_variance()[2]*100
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree,c="red",marker='o')
    plt.xlabel("Number of Principal Components")
    plt.ylabel("Percentage Explained Variance")
    plt.title("Scree Plot")
    plt.show(block=False)
```

```python
print('Proportional variance:\n{0}'.format(pca.get_factor_variance()[1]))
print('Cumulative variance:\n{0}'.format(pca.get_factor_variance()[2]))

display_scree_plot(pca) 
#we need 7 variables to have 80.19% of the variance explained
```

```python
pcarotated = FactorAnalyzer(n_factors = 7, rotation = 'Varimax',method='principal')
pcarotated.fit(forpca)
loadings = pd.DataFrame(pcarotated.loadings_, columns=['PC1', 'PC2','PC3','PC4',
                                                       'PC5','PC6','PC7'], index=forpca.columns)
loadings[(loadings>0.3)|(loadings<-0.3)]
```

```python
#just making sure
print(datanum[['Income','MntMeatProducts','NumCatalogPurchases','NumWebVisitsMonth']].corr())
print(datanum[['Kidhome','MntWines','NumWebPurchases','NumStorePurchases','NumCatalogPurchases']].corr())
```

```python
reduced = pd.DataFrame(pcarotated.transform(datanum),
                       columns = ['Wealthiness','AmtDailyNeeds','PurchFreq',
                                  'Relative Age','Deal_Hunting_Tendencies',
                                  'Timescalesincemember','AmtGolds'])
forclust = pd.concat([reduced.reset_index(drop=True),datacat.reset_index(drop=True)],axis = 1)
```

```python
# Function for plotting elbow curve
from kmodes.kprototypes import KPrototypes
Poscatcol = [forclust.columns.get_loc(i) for i in forclust.loc[:, forclust.dtypes == 'object'].columns]
cost = []
for i in range(2,7):
    clust = KPrototypes(n_clusters = i,init='Huang',random_state=42)
    clust.fit_predict(forclust,categorical = Poscatcol)
    cost.append(clust.cost_)
    
sns.lineplot(x = range(2,7),y = cost,marker = "+",dashes = True)
plt.title('Elbow Curve')
plt.xlabel('N of Clusters')
plt.ylabel('Cost')
plt.plot
```

```python
#I decide to cluster the data into 3 cluster based on the Elbow Curve.
#more cluster might be better but sometimes we don't want to group our customers into many segments since it will also
#give ourselves a hard time.
finalclust = KPrototypes(n_clusters = 3, init='Huang',random_state=42)
finalclust.fit_predict(forclust,categorical = Poscatcol)
full = forclust.loc[:]
full['segments'] = finalclust.labels_
full.segments.value_counts()
```

```python
#This is if we plot the latent variables
plt.style.use('ggplot') 
fig, ax = plt.subplots(nrows = 2, ncols = 4,figsize = (15,7))
sns.boxplot(y='Wealthiness', x= 'segments', data=full, ax=ax[0,0])
sns.boxplot(y='AmtDailyNeeds', x= 'segments', data=full, ax=ax[0,1])
sns.boxplot(y='PurchFreq', x= 'segments', data=full, ax=ax[0,2])
sns.boxplot(y='Relative Age', x= 'segments', data=full, ax=ax[0,3])
sns.boxplot(y='Deal_Hunting_Tendencies', x= 'segments', data=full, ax=ax[1,0])
sns.boxplot(y='Timescalesincemember', x= 'segments', data=full, ax=ax[1,1])
sns.boxplot(y='AmtGolds', x= 'segments', data=full, ax=ax[1,2])
fig.delaxes(ax[1,3])
plt.tight_layout()
plt.show()
```
