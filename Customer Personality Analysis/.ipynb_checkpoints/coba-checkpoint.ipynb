```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rawdata = pd.read_csv('model and plot.csv')
rawdata.segments.value_counts()
```

```python
modelinput = rawdata.drop(['Unnamed: 0'],axis = 1,inplace = False)
modelinput.head()
```

```python
#we split the data to train and test to see how well does the model learn from the data
from sklearn.model_selection import train_test_split
x = modelinput.drop('segments',axis = 1)
y = modelinput.segments

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.35,random_state =42)
```

```python
#This is only one of many approaches available out there. I want to scale the numeric variables but excluding 
#the categorical data, leaving the categorical data the way it is

from sklearn.preprocessing import StandardScaler
def preprocess(feat, train = True):
    
    featnum = feat.loc[:,feat.dtypes != 'object']
    featcat = feat.loc[:,feat.dtypes == 'object']
    #We consider the PhD to be the highest education since usually it's 
    #professors who took this degree.
    edudict = {'Basic' : 0,'2n Cycle':1,'Graduation':2,'Master':3,'PhD':4}
    featcat = featcat.replace({'Yes':1,'No':0},regex=True)
    featcat['Education'] = featcat['Education'].replace(edudict)
    catwithdum = pd.get_dummies(featcat, columns=["Marital_Status"])
    catwithdum = catwithdum.astype('int')
    
    if train == True:
        featnum = pd.DataFrame(scale.fit_transform(featnum),columns = featnum.columns)
    elif train == False:
        featnum = pd.DataFrame(scale.transform(featnum),columns = featnum.columns)
    
    return pd.concat([featnum.reset_index(drop=True),catwithdum.reset_index(drop=True)],axis = 1)  
```

```python
scale = StandardScaler()
xtrainclean = preprocess(xtrain,train=True)

from sklearn.metrics import classification_report
from xgboost import XGBClassifier,plot_importance
model = XGBClassifier(booster = 'gbtree',
                      objective = 'multi:softmax',
                      eval_metric = 'mlogloss',
                      seed = 42,
                      use_label_encoder = False,
                      num_parallel_tree = 100)
model.fit(xtrainclean,ytrain)


xtestclean = preprocess(xtest,train=False)
pred = model.predict(xtestclean)

print(classification_report(ytest,pred))
```

The XGBoost is indeed very strong (hahaha XD)

Now, we'll try to see which variables are important to be interpreted

```python
plot_importance(model,max_num_features = 10)
plt.show()
```

Okay we clearly see that Income is easily used to differentiate each cluster. What about if we don't use Income?

```python
xtrain_noincome = preprocess(xtrain.drop('Income',axis = 1),train = True)
model2 = XGBClassifier(booster = 'gbtree',
                      objective = 'multi:softmax',
                      eval_metric = 'mlogloss',
                      seed = 42,
                      use_label_encoder = False,
                      num_parallel_tree = 100)
model2.fit(xtrain_noincome,ytrain)


xtest_noincome = preprocess(xtest.drop('Income',axis =1),train=False)
pred2 = model2.predict(xtest_noincome)

print(classification_report(ytest,pred2))
plot_importance(model2,max_num_features = 10)
plt.show()
```
